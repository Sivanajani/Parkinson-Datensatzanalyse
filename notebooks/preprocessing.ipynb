### Datentyp und Numerische Codes in kategoriale Typen umzuwandeln (Geschlecht)

#### subject# (Teilnehmer-ID)
"category" anstatt int64, da diese Spalte verwendet wird, um Teilnehmer eindeutig zu kennzeichnen und nicht für mathematische Operationen verwendet wird. Der Speicher wird optimiert und unangemessene Berechnungen werden verhindert.

```python
df['subject#'] = df['subject#'].astype('category')
```

```python
# Anzahl Teilnehmer:
unique_participants = df['subject#'].nunique()
print(f"Anzahl der eindeutigen Teilnehmer: {unique_participants}")
```

#### sex (Geschlecht)
Braucht keine Datentypumwandlung, da der Datentyp bereits "category" ist. Es gibt nur wenige eindeutige Werte und keine arithmetischen Operationen werden benötigt. Aber Nummerischer Code muss in kategorialer Typ umgewandelt werden!

```python
df['sex'] = df['sex'].astype('category')
```

```python
print(df.dtypes)
```

```python
#Kontrollieren ob Spalte "sex" 0 und 1 Werte hat oder andere die nicht korrekt sind --> "0" = männlich, "1" = weiblich
df.sex.value_counts()
```

```python
# 0 und 1 nach M und F ersetzen, da es sonst einen Mittelwert gibt, den wir in der Statistik nicht wollen
df['sex'] = df['sex'].replace({0: 'M', 1: 'F'})

# Überprüfung der Konvertierung
print(df['sex'].value_counts())
```

```python
# Anzahl eindeutigen männlichen und weiblichen Teilnehmer, unter der Annahme, dass 'subject#' pro Teilnehmer eindeutig ist
unique_sex_counts = df.drop_duplicates(subset='subject#')['sex'].value_counts()
print("Anzahl der eindeutigen männlichen (M) und weiblichen (F) Teilnehmer:")
print(unique_sex_counts)
```

#### test_time (Testzeit)
Testzeit wird von Float zu Integer gecastet, um eine Gruppierung der Aufnahmesitzungen zu ermöglichen.

```python
df['test_time'] = df['test_time'].astype('int64')
```

```python
print(df.dtypes)
```

#### Statistische Methoden: Interquartilsabstand (IQR) verwenden, um Abweichungen zu erkennen.

```python
Q1 = df['Jitter(%)'].quantile(0.25)
Q3 = df['Jitter(%)'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filtere Abweichungen
outliers = df[(df['Jitter(%)'] < lower_bound) | (df['Jitter(%)'] > upper_bound)]
print(outliers)
```

### Fehlerhafte Einträgen und Behebung

#### Negative und 0-Werte im Datensatz

```python
#schauen ob im Datensatz negative oder 0 Werte vorhanden sind
numeric_df = df.select_dtypes(include=[np.number])

has_negative_and_0_values = (numeric_df <= 0).any().any()

print(has_negative_values)
```

```python
# Einschränkung auf numerische Daten, falls notwendig
numerical_df = df.select_dtypes(include=[np.number])
zero_counts = (numerical_df <= 0).sum()

print("Anzahl der negativen- und 0-Werten in jeder numerischen Spalte:")
print(zero_counts)
```

```python
# Filtern der Werte, bei denen 'test_time' zwischen 0 und 1 liegt, da weniger als ein Tag nicht realistisch
filtered_values = df[(df['test_time'] >= 0) & (df['test_time'] <= 1)]

# Anzeigen dieser gefilterten Werte
print(filtered_values['test_time'])
```

```python
#durchschnittswerte jeder Spalte pro Teilnehmer
for subject_num in df['subject#'].unique():
    df_subject = df[df['subject#'] == subject_num]
    numerical_columns = df_subject.select_dtypes(include=[np.number]).columns.tolist()
    mean_values = df_subject[numerical_columns].mean()
    print("Mean values for Subject", subject_num, ":")
    print(mean_values)
    print("\n")
```

#### test_time (Testzeit) 

```python
negative_test_time = df[df['test_time'] <= 0]

if negative_test_time.empty:
    print("Keine falsche Testzeit gefunden.")
else:
    print(f"Falsche Testzeit:\n{negative_test_time.test_time}")
```

```python
# Zeigt uns welche Teilnehmer die Miuszeiten und 0 hatten
negative_times_df = df[df['test_time'] <= 0]
negative_times_count = negative_times_df.groupby('subject#').size()
negative_times_count = negative_times_count[negative_times_count > 0]

if negative_times_count.empty:
    print("Keine Teilnehmer mit negativen Testzeiten gefunden.")
else:
    print("Teilnehmer und ihre Anzahl negativer Testzeiten:")
    print(negative_times_count)
```

```python
# Zeigt uns welche Teilnehmer die Miuszeiten hatten
negative_times_df = df[df['test_time'] <= 0]
negative_times_count = negative_times_df.groupby('subject#').size()
negative_times_count = negative_times_count[negative_times_count > 0]

if negative_times_count.empty:
    print("Keine Teilnehmer mit negativen oder 0 Testzeiten gefunden.")
else:
    print("Teilnehmer und ihre Anzahl negativer Testzeiten:")
    print(negative_times_count)
```

```python
#NaN Werte von test_time ausgeben
nan_values_test_time = df[df['test_time'].isnull()]['test_time']
print(nan_values_test_time)
```

```python
# Anzahl eindeutigen männlichen und weiblichen Teilnehmer, unter der Annahme, dass 'subject#' pro Teilnehmer eindeutig ist
unique_sex_counts = df.drop_duplicates(subset='subject#')['sex'].value_counts()
print("Anzahl der eindeutigen männlichen (M) und weiblichen (F) Teilnehmer:")
print(unique_sex_counts)
```

```python
# Gruppieren der Daten nach 'subject#' und 'test_time', dann Zählen der Messungen in jeder Gruppe
measurements_per_test_time = df.groupby(['subject#', 'test_time']).size()

# Umwandeln in DataFrame für bessere Handhabbarkeit
measurements_per_test_time = measurements_per_test_time.reset_index(name='count')

# Berechnen der durchschnittlichen Anzahl von Messungen pro 'test_time' pro 'subject#'
average_measurements_per_test_time = measurements_per_test_time.groupby('subject#')['count'].mean()

# Berechnen der durchschnittlichen Anzahl von verschiedenen 'test_time' pro 'subject#'
average_different_test_times_per_subject = df.groupby('subject#')['test_time'].nunique()

# Gesamtdurchschnitt über alle Testpersonen für die Anzahl von Messungen pro 'test_time'
overall_average_measurements = average_measurements_per_test_time.mean()

# Gesamtdurchschnitt über alle Testpersonen für die Anzahl von verschiedenen 'test_time'
overall_average_different_test_times = average_different_test_times_per_subject.mean()

# Ausgabe der Ergebnisse
print("Durchschnittliche Anzahl von Messungen pro 'test_time' pro Testperson:", overall_average_measurements)
print("Durchschnittliche Anzahl von verschiedenen 'test_time' pro Testperson:", overall_average_different_test_times)
```

```python
# Durchschnitt der Beobachtungen pro Testperson ohne NAN-Werte
average_observations = observations_per_subject.mean()

# Berechnet den Durchschnitt der Beobachtungen pro Testperson und konvertiere das Ergebnis in eine Ganzzahl
average_observations_int = int(round(observations_per_subject.mean()))

print("\nDurchschnittliche Anzahl der Messpunkte pro Testperson:")
print(average_observations)

print("\nDurchschnittliche Anzahl der Messpunkte pro Testperson, als ganze Zahl:")
print(average_observations_int)
```

### Funktionen für Visualisierung
Erstellen neuer Funktionen für Analysen.

#### Interaktionsmerkmale
Aufzeigen wie sich verschiedene Symptome (z. B. motorische Symptome und Sprachmessungen wie Jitter und Shimmer) gegenseitig beeinflussen. Die Korrelation von motorischen UPDRS-Werten mit Stimm-Tremordaten könnte zeigen, wie die Schwere der motorischen Symptome Sprachstörungen beeinflusst.

```python
df['motor_Jitter_interaction'] = df['motor_UPDRS'] * df['Jitter(%)']
df['motor_Shimmer_interaction'] = df['motor_UPDRS'] * df['Shimmer']
```

```python
# Überprüfung die an der Berechnung beteiligten Spalten und die neuen Spalten
print(df[['motor_UPDRS', 'Jitter(%)', 'motor_Jitter_interaction', 'Shimmer', 'motor_Shimmer_interaction']].head())
```

```python
# Beschreibende Statistiken für die neuen Interaktionsfunktionen
print(df[['motor_Jitter_interaction', 'motor_Shimmer_interaction']].describe())
```

```python
# Überprüfung für negative Werte
if (df['motor_Jitter_interaction'] < 0).any() or (df['motor_Shimmer_interaction'] < 0).any():
    print("Unexpected negative values found in interaction features.")

# Auf NaN-Werte prüfen
print(df[['motor_Jitter_interaction', 'motor_Shimmer_interaction']].isnull().sum())
```
